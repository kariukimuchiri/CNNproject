```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=df, x='Wavelength (nm)', y='Abs', marker="+")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=df1, x='Wavelength(nm)', y='Abs', marker="+")
```

```python
plt.figure(figsize=(15,8))
sns.kdeplot(data=df, x="Abs")
# distribution plot of Abs in 30% ethanol soln
```

```python
fig=px.scatter(df, x='Wavelength(nm)', y='Abs', title='30% ethanol soln graph of Wavelength against Abs')  
fig.show()
fig.write_html('30% ethanol soln graph of Wavelength against Abs.html', auto_open=True)
#Line graph showing the wavelength plotted against Abs
```

```python
fig=px.histogram(df, x='Abs', title='30% ethanol soln Distribution Count of Abs')
fig.show()
fig.write_html('30% ethanol soln .html', auto_open=True)
#Shows the count of Abs and its distribution
```

```python
fig=px.box(df,x='Abs', title='30% ethanol soln Box plot of Abs')
fig.show()
fig.write_html('30% ethanol soln Box plot of Abs.html', auto_open=True)
#This shows the max, min, median, 1st, 2nd and 3rd quartile
#You can also see the outliers at 10 by pointing at the graph
```

```python
fig=px.scatter(df1, x='Wavelength(nm)', y='Abs', title='dye + washing soln against Abs')  
fig.show()
fig.write_html('dye + washing soln against Abs.html', auto_open=True)
#Line graph showing the wavelength plotted against Abs
```

```python
fig=px.histogram(df1, x='Abs', title='Distribution count of Abs')
fig.show()
fig.write_html('Distribution count of Abs.html', auto_open=True)
#Shows the count of Abs and its distribution
```

```python
fig=px.box(df1,x='Abs', title='Box plot of Abs')
fig.show()
fig.write_html('Box plot of Abs.html', auto_open=True)
```
